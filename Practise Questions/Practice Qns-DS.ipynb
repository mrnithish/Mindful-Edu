1- Given an array of 1s and 0s which has all 1s first followed by all 0s. Find the number of 0s.

 Count the number of zeroes in the given array. Please try to solve it in O(log(n)).

```python
def count_zeros(arr):
    left,right=0,len(arr)-1
    while left < right:
        mid=(left+right)//2
        if arr[mid]==1:
            left=mid+1
        else:
            right=mid
        
        if left<len(arr) and arr[left]==0:
            return len(arr)-left

# Example usage:
array = [0,0,1]
array.sort(reverse=True)
print(array)
print(count_zeros(array))  # Output: 4
```

2-  Given an integer X, find its square root. If X is not a perfect square, then return floor(âˆšx). Do it in O(log(n)) time.

```python
def floor_sqrt(x):
    left=0
    right=x

    while left<right:
        mid=(left+right)//2
        if mid*mid < x:
            left=mid+1
        elif mid*mid > x:
            right=mid-1
        else:
            return mid
    return right

# Example usage:
x = 0
print(floor_sqrt(x))  # Output: 4
```

3-  Reverse a Queue using Recursion.

```python
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        newnode = Node(data)
        if self.rear is None:
            self.front = self.rear = newnode
        else:
            self.rear.next = newnode
            self.rear = newnode
    
    def dequeue(self):
        if self.front is None:
            print("Queue is Empty")
            return None
        else:
            temp = self.front
            self.front = self.front.next
            if self.front is None:
                self.rear = None
            return temp.data
    
    def display(self):
        if self.front is None:
            print("Queue is Empty")
        else:
            temp = self.front
            while temp:
                print(temp.data, end="->")
                temp = temp.next
            print("None")

    def reverse(self):
        if self.front is not None:
            self._reverse_recursive()

    def _reverse_recursive(self):
        if self.front is None:
            return
        front_data = self.dequeue()
        self._reverse_recursive()
        self.enqueue(front_data)

# Example usage:
o = Queue()
o.enqueue(3)
o.enqueue(5)
o.enqueue(52)
print("Original queue:")
o.display()

o.reverse()
print("Reversed queue:")
o.display()
```

 4- You are given two strings s and t.


String t is generated by random shuffling string s and then add one more letter at a random position.



Return the letter that was added to t

```python
def find_extra_letter(s: str, t: str) -> str:
    # Create a frequency dictionary for string s
    freq_s = {}
    for char in s:
        if char in freq_s:
            freq_s[char] += 1
        else:
            freq_s[char] = 1

    # Create a frequency dictionary for string t
    freq_t = {}
    for char in t:
        if char in freq_t:
            freq_t[char] += 1
        else:
            freq_t[char] = 1

    # Find the character that has a different frequency in t
    for char in freq_t:
        if char not in freq_s or freq_t[char] > freq_s[char]:
            return char

    # If no extra character is found, return an empty string
    return ""

s="cat"
t="ceat"
print(find_extra_letter(s,t))
```
