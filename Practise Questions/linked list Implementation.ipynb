# Doubly Linked List

```python
#Doubly Linked List
class Node:
    def __init__(self,data) -> None:
        self.prev=None
        self.data=data 
        self.next=None

class DoublyLinkedList:
    def __init__(self) -> None:
        self.head=None
    
    def display(self):
        node=self.head
        if node is None:
            print("Empty")
        else:
            while node is not None:
                print(node.data,end="->")
                node=node.next
            print()


    def reverse(self):
        node=self.head
        if node is None:
            print("empty")
        else:
            while node.next is not None:
                node=node.next
            while node is not None:
                print(node.data,end="->")
                node=node.prev
            print()


    def insertAtEnd(self,data):
        newnode=Node(data)
        if self.head is None:
            self.head=newnode
        else:
            node=self.head
            while node.next is not None:
                node=node.next
            node.next=newnode
            newnode.prev=node
    
    def insertAtBegin(self,data):
        newnode=Node(data)
        if self.head is None:
            self.head=newnode
        else:
            self.head.prev=newnode
            newnode.next=self.head
            self.head=newnode
    
    def insertafter(self,data,x):
        if self.head is None:
            print("Empty")
        else:
            node=self.head
            newnode=Node(data)
            while node is not None:
                if node.data==x:
                    break
                node=node.next
            if node is None:
                print("Element not Found")
            else:
                newnode.next=node.next
                newnode.prev=node
                if node.next is not None:
                    node.next.prev=newnode
                node.next=newnode
    
    def insertbefore(self,data,x):
        if self.head is None:
            print("Empty")
        else:
            node=self.head
            newnode=Node(data)
            while node is not None:
                if node.data==x:
                    break
                node=node.next
            if node is None:
                print("Element not Found")
            else:
                newnode.next=node
                newnode.prev=node.prev
                if node.prev is not None:
                    node.prev.next=newnode
                else:
                    self.head=newnode
                node.prev=newnode
    
    def deleteAtBegin(self):
        if self.head is None:
            print("DDL is Empty")
        elif self.head.next is None:
            self.head=None
            print("Last Element Deleted")
        else:
            self.head=self.head.next
            self.head.prev=None
    
    def deleteAtEnd(self):
        if self.head is None:
            print("DDL is Empty")
        elif self.head.next is None:
            self.head=None
            print("Last Element Deleted")
        else:
            node=self.head
            while node.next:
                node=node.next
            node.prev.next=None

    def deleteByValue(self,x):
        if self.head is None:
            print("DDL is Empty")
        if self.head.next is None:
            if x==self.head.data:
                self.head=None
            else:
                print("Element not presented")
        if self.head.data ==x:
            self.head=self.head.next
            self.prev=None
        node=self.head
        while node.next:
            if node.data ==x:
                break
            node=node.next
        if node.next :
            node.next.prev =node.prev
            node.prev.next=node.next
        else:
            if node.data==x:
                node.prev.next=None
            else:
                print("Element not Present")
      
            


ob=DoublyLinkedList()
ob.insertAtEnd(3)
ob.insertAtEnd(32)
ob.insertAtBegin(1)
ob.insertAtBegin(8)
ob.display()
ob.reverse()
ob.insertafter(2,1)
ob.display()
ob.reverse()
ob.insertbefore(0,1)
ob.display()
ob.reverse()
ob.deleteAtBegin()
ob.display()
ob.reverse()
ob.deleteAtEnd()
ob.display()
ob.reverse()
ob.deleteByValue(2)
ob.display()
ob.reverse()
```

```python
# Singly Linked List
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

# Linked List 
class LinkedList:

    #Head initializing
    def __init__(self):
        self.head=None

    #Insert Begin
    def insertAtBegin(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            new_node.next=self.head
            self.head=new_node

    #Insert At End
    def insertAtEnd(self,data):
        new_node=Node(data)
        node=self.head
        if self.head  is  None:
            self.head=new_node
        else:
            while node.next is not None: # It needs to be stopped at the last So use node.next
                node=node.next
            node.next=new_node

    #Insert the element after the given value
    def insert_after(self,data,x):
        node=self.head
        while node  is not None:
            if x==node.data:
                break
            else:
                node=node.next
        if node  is None:
            print("The Given Value is not Present in the LinkedList")
        else:
            new_node=Node(data)
            new_node.next=node.next
            node.next=new_node
    

    #Insert the element before the given value
    def insert_before(self,data,x):
        new_node=Node(data)
        node=self.head
        if node is None:
            print("List is Empty")
        if node.data == x:
            new_node.next=node
            node=new_node
        else:
            while node.next is not None:
                if node.next.data==x:
                    break
                node=node.next
            if node.next is None:
                print("Not Exist")
            else:
                new_node.next=node.next #Similar insert after element logic
                node.next=new_node
    
    #Insert at the Postion
    def insert_position(self,data,pos):
        new_node=Node(data)
        node=self.head
        if node is None:
            print("List is Empty")
        if pos==0:
            new_node.next=node
            node=new_node
        else:
            count=1
            while node is not None:
                if count==pos: # It places the specific postion of index
                    break
                count+=1
                node=node.next
            if node is None:
                print("Index Not Found")
            else:
                new_node.next=node.next
                node.next=new_node

    #Delete at Begin
    def deleteAtBegin(self):
        if self.head is None:
            print("List is Empty")
        else:
            self.head=self.head.next
    
    #Delete at the End
    def deleteAtEnd(self):
        if self.head is None:
            print("List is Empty")
        elif self.head.next is None:
            self.head=None
        else:
            node=self.head
            while node.next.next is not None:
                node=node.next
            node.next=None

    #delete by value
    def deleteByValue(self,value):
        node=self.head
        if node is None:
            print("List is Empty")
        elif node.data is value:
            self.head=self.head.next
        else:
            while node.next is not None:
                if node.next.data== value:
                    break
                node=node.next
            if node.next is None:
                print("Not present in the List")
            else:
                node.next=node.next.next

    #Searching
    def searchElement(self,value):
        node=self.head
        if self.head is None:
            print("List is Empty")
        elif node.data == value:
            print("Element Found at 0th position")
        else:
            pos=1
            while node.next != None:
                if node.data==value:
                    break
                pos+=1
                node=node.next
            if node.next is None:
                print("Element Not Present")
            else:
                print(f'{value} is found at {pos}th position')

    #Get the Middle  Element in a list
    def getMiddle(self):
        node=self.head
        slow,fast=node,node.next
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
        print(slow.data)
    
    #reverse
    def reverseList(self,curr,prev):
        if curr.next is None:
            self.head=curr
            curr.next=prev
            return
        temp=curr.next
        curr.next=prev
        self.reverseList(temp,curr)



    def reverse(self):
        if self.head is None:
            print("List is Empty")
            return
        self.reverseList(self.head,None)




    # Traversal
    def display(self):
        node=self.head
        if node is None:
            print("List is Empty")
        else:
            while node is not None: # We need to traverse via the last node
                print(node.data,end=" -> ")
                node=node.next
            print()


obj=LinkedList()
obj.insertAtEnd(0)
obj.insertAtBegin(10)
obj.insertAtBegin(20)
obj.insertAtEnd(34)
obj.insert_after(400,34)
obj.insert_before(4,10)
obj.insert_position(100,2)
obj.display()
obj.deleteAtBegin()
obj.display()
obj.deleteAtEnd()
obj.display()
obj.searchElement(10)
obj.getMiddle()
obj.display()
obj.reverse()
obj.display()
```

# Queue

```python
# Queue Linked List
class Node:
    def __init__(self,data) -> None:
        self.data=data 
        self.next=None

class Queue:
    def __init__(self):
        self.front=None
        self.rear=None

    def enqueue(self,data):
        newnode=Node(data)
        if self.front is None:
            self.rear=newnode
            self.front=newnode
        else:
            self.rear.next=newnode
            self.rear=newnode
    
    def dequeue(self):
        if self.front is None:
            print("Queue is Empty")
        elif self.front.next is None:
            print("pop")
            self.front=None
        else:
            self.front=self.front.next
    
    def display(self):
        if self.front is None:
            print("Queue is Empty")
        else:
            temp=self.front
            while temp :
                print(temp.data, end="->")
                temp=temp.next


o=Queue()
o.enqueue(3)
o.enqueue(5)
o.enqueue(52)
o.display()
            

```

```python
#queue
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        newnode = Node(data)
        if self.rear is None:
            self.front = self.rear = newnode
        else:
            self.rear.next = newnode
            self.rear = newnode
    
    def dequeue(self):
        if self.front is None:
            print("Queue is Empty")
            return None
        else:
            temp = self.front
            self.front = self.front.next
            if self.front is None:
                self.rear = None
            return temp.data
    
    def display(self):
        if self.front is None:
            print("Queue is Empty")
        else:
            temp = self.front
            while temp:
                print(temp.data, end="->")
                temp = temp.next
            print("None")

    def reverse(self):
        if self.front is not None:
            self._reverse_recursive()

    def _reverse_recursive(self):
        if self.front is None:
            return
        front_data = self.dequeue()
        self._reverse_recursive()
        self.enqueue(front_data)

# Example usage:
o = Queue()
o.enqueue(3)
o.enqueue(5)
o.enqueue(52)
print("Original queue:")
o.display()

o.reverse()
print("Reversed queue:")
o.display()
```

# Stack

```python
# Stack Linked List
class Node:
    def __init__(self,data):
        self.data=data 
        self.next=None

class Stack:
    def __init__(self):
        self.top=None

    def push(self,data):
        newnode=Node(data)
        if self.top is None:
            self.top=newnode
            self.top.next=None
        else:
            newnode.next=self.top
            self.top=newnode
            
    def pop(self):
        if self.top is None:
            print("empty")
        elif self.top.next is None:
            print("pop",self.top.data)
            self.top=None
        else:
            temp=self.top
            print("pop",self.top.data)
            self.top=temp.next
            temp=None
    
    def display(self):
        if self.top is None:
            print("empty")
        else:
            temp=self.top
            while temp:
                print(temp.data,end="->")
                temp=temp.next

o=Stack()
o.push(2)
o.push(3)
o.pop()
o.display()
```

```python
#Stack
class Node:
    def __init__(self, data):
        self.data = data 
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        newnode = Node(data)
        if self.top is None:
            self.top = newnode
        else:
            newnode.next = self.top
            self.top = newnode
            
    def pop(self):
        if self.top is None:
            print("empty")
            return None
        else:
            temp = self.top
            self.top = self.top.next
            popped_data = temp.data
            temp = None
            return popped_data
    
    def display(self):
        if self.top is None:
            print("empty")
        else:
            temp = self.top
            while temp:
                print(temp.data, end="->")
                temp = temp.next
            print("None")

    def reverse(self):
        if self.top is not None:
            self._reverse_recursive()

    def _reverse_recursive(self):
        if self.top is None:
            return
        
        top_data = self.pop()
        self._reverse_recursive()
        
        self._insert_at_bottom(top_data)
    
    def _insert_at_bottom(self, data):
        if self.top is None:
            self.push(data)
        else:
            top_data = self.pop()
            self._insert_at_bottom(data)
            self.push(top_data)

o = Stack()
o.push(2)
o.push(3)
o.push(5)
print("Original stack:")
o.display()

o.reverse()
print("Reversed stack:")
o.display()
```
